```python
import numpy as np
import cv2
```

```python
# loading the files from the file1 
with_mask=np.load('with_mask.npy')
without_mask=np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
# converting the 4d array into 2d array
with_mask=with_mask.reshape(200,50*50*3)
without_mask=without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape
# has 200 rows and 7500 columns
```

```python
without_mask.shape
# has 200 rows and 7500 columns
```

```python
# concatining the rows
X=np.r_[with_mask,without_mask]
```

```python
X.shape
# first 200 rows with mask and the rest 200 data without mask
```

```python
# for column making everyone 0 as we have 400 zeroes
labels=np.zeros(X.shape[0])
```

```python
# first 200 will be zeroes and the rest will be 1 i.e. 0->with_mask 1->without_mask
labels[200:]=1.0
```

```python
names={0:'Mask',1:'NoMask'}
```

```python
# svm=support vector machine
# svc=support vector classifier
from sklearn.svm import SVC

# to check the accuracy 
from sklearn.metrics import accuracy_score
```

```python
# will be used to divide our data into train and test part,training part will be used to apply the ml algo and test part will be used to check the accuracy
from sklearn.model_selection import train_test_split
```

```python
# we are seting (.25) i.e. 25 data for testing and 75 percent data for training
x_train,x_test,y_train,y_test=train_test_split(X,labels,test_size=0.25)
```

```python
x_train.shape
```

```python
# its used to reduce the dimentions 
# pca=principal component analysis
from sklearn.decomposition import PCA
```

```python
# we are giving n_components =3 so our data is converted into 3 dimentions 
pca=PCA(n_components=3)
x_train=pca.fit_transform(x_train)

# the maths behind this dimentionality redction use eigen values and eigen vectors
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
svm=SVC()
svm.fit(x_train,y_train)
```

```python
x_test=pca.transform(x_test)
y_pred=svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
haar_data=cv2.CascadeClassifier('data.xml')
capture=cv2.VideoCapture(0)
data=[]
font = cv2.FONT_HERSHEY_COMPLEX       #assiging the font 
while True:
    flag,img=capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,102,102),4)
            face=img[y:y+h, x:x+w,: ]
            face=cv2.resize(face,(50,50))
            face=face.reshape(1,-1)
            face=pca.transform(face)
            pred=svm.predict(face)[0]   #return 0 or 1
            n=names[int(pred)]
            cv2.putText(img,n,(x,y),font,1,(244,250,250),2)    # puting the text on the screen whether the person is wearing a mask or not
            print(n)
        cv2.imshow('Window',img)
        if cv2.waitKey(2)==27 : #27 is the ASCII val of Esc key so if we press Esc the window will close
            break
capture.release()
cv2.destroyAllWindows() 
```
